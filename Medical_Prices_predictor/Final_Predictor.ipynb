```python
# Importing the important modules

import matplotlib
import random
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
# Code for creating the data file
#Prices represent the price of medical bill 
# whereas the bill_no represent the bill number respective to its price.



prices=['Prices']
# Generating the prices of the medical bills
for i in range(2000):
    prices.append(random.randint(100,1000))

Bill_no=list(range(1,2001))   

# Writing the data into the csv file
import csv
csvfile='prices.csv'
with open(csvfile,'w') as output:
    writer=csv.writer(output,lineterminator='\n')
    for price in prices:
        writer.writerow([price])


dataFrame = pd.read_csv("prices.csv")
dataFrame['Bill_no'] = Bill_no
dataFrame
```

```python
# Reading the data from the file
Bill_price = dataFrame['Prices']
Bill_number = dataFrame['Bill_no']
 
Bill_number = Bill_number.reshape(len(Bill_number),1)
Bill_price = Bill_price.reshape(len(Bill_price),1)
```

```python
# Splitting the data into the testing and training data
# Since I had 2000 enteries in my file so first 1750 enteries are used as training data and rest 250 are used as 
# testing data
Y_train = Bill_price[:1750]
Y_test = Bill_price[1750:]
X_train = Bill_number[:1750]
X_test = Bill_number[1750:]
```

```python
# Plotting the data
# Data is plotted to get some insight from visualising the data, Sometmes by looking at data we can 
# find some of the hidden patterns in the data
plt.scatter(X_test, Y_test,  color='black')
plt.title('Test Data')
plt.xlabel('Bill_Number')
plt.ylabel('Price')
plt.xticks(())
plt.yticks(())
 
plt.show()
```

```python
# Create linear regression object
regObject = linear_model.LinearRegression()
 
# Train the model using the training sets
regObject.fit(X_train, Y_train)
 
# Plot outputs
plt.plot(X_test, regObject.predict(X_test), color='red',linewidth=2)
```
